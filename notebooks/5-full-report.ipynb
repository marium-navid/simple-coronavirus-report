```python
import pandas as pd

import chart_utils
import utils

from IPython.display import Markdown, HTML

# Default parameters
fulldate_format = utils.fulldate_format
```

```python
# Import data
df = utils.clean_jhu()
```

```python
def add_caption(df, county):
    subset_df = df[df.county==county]
    
    COUNTY_NAME = subset_df.county.iloc[0]
    MOST_RECENT_DATE = subset_df.date.max().strftime(fulldate_format)
    
    most_recent_df = subset_df[subset_df.date == pd.to_datetime(MOST_RECENT_DATE)]
    MOST_RECENT_CASES = most_recent_df.cases.iloc[0]
    MOST_RECENT_DEATHS = most_recent_df.deaths.iloc[0]
    
    NEW_CASES = most_recent_df.cases_avg7.iloc[0]
    
    display(Markdown(f"## {COUNTY_NAME}"))
    display(
        Markdown(
            f"As of {MOST_RECENT_DATE}, there were {MOST_RECENT_CASES:,} cases and {MOST_RECENT_DEATHS:,} deaths. "
            f"The 7-rolling average for new cases is {NEW_CASES:,.1f}."
        )
    )
```

```python
# Make charts
plot_me = ["Los Angeles", "Riverside", "San Bernardino", 
          "Imperial", "San Diego"]

START_DATE = "8/1/20"

for c in plot_me: 
    add_caption(df, c)
    chart_utils.make_chart(df, c, START_DATE)
```
